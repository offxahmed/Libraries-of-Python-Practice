```python
# stride in the python 
# the strides attribute of an array (arr.strides) returns a tuple of
# integers representing the byte offsets (strides) 
# needed to step from one element to the next along each dimension of the array.
import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])  # Shape: (2, 3), dtype: int64 (8 bytes/item)
print(arr.strides)  # Output: (24, 8)

# 24 (row stride): To go to the next row, skip 3 elements × 8 bytes = 24 bytes.
# 8 (col stride): To go to the next column in the same row, skip 1 element × 8 bytes = 8 bytes.
```

```python
# Numpy vs List 
a=[i for i in range(10000000)]
b=[i for i in range(10000000,20000000)]
c=[]
import time
start=time.time()
for i in range (len(a)):
    c.append(a[i]+b[i])

print(time.time()-start)
    
    
```

```python
# Advance Indexing
# Fancy Indexing
# extract 1 2 and 4th row
#extract the 2nd and 4th column
a=np.arange(24).reshape(6,4)
print(a)
print(a[[0,2,3]])
print(a[:,[1,3]])

```

```python
# Broad casting : The term broadcasting describes how NumPy treats arrays with different shapes
# during arithmetic operations.
#The smaller array is "broadcast- across the larger array so that they have compatible shapes.
a=np.arange(6).reshape(2,3)
b=np.arange(3,6).reshape(1,3)
print(a)
print(b)
print(a+b)
```

```python
a=np.arange(12).reshape(4,3)
b=np.arange(6,9).reshape(3)
print(a)
print(b)
print(a+b)
```

```python
a=np.arange(12).reshape(3,4)
b=np.arange(6,9).reshape(3)
print(a)
print(b)
print(a+b)
```

```python
a=np.arange(3).reshape(1,3)
b=np.arange(6,9).reshape(3,1)
print(a)
print(b)
print(a+b)
```

```python
a=np.arange(3).reshape(3,1)
b=np.arange(6,9).reshape(3)
print(a)
print(b)
print(a+b)
```

```python
# mean squared  error
actual=np.random.randint(1,50,25)
predicted=np.random.randint(1,50,25)
print(actual)
print(predicted)
def MSE(actual,predicted):
    return np.mean((actual-predicted)**2)

MSE(actual,predicted)
    
```

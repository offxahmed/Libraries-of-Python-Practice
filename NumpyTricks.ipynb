```python
# Contatination on the Np Array 
c=np.arange(1,100,6).reshape(2,3)
d=np.arange(6,12).reshape(2,3)
print(c)
print(d)
```

```python
row_cont=np.concatenate((c,d),axis=0)
col_cont=np.concatenate((c,d),axis=1)
print(row_cont)
print(col_cont)
```
